# Workforce Compensation Planning

```python
import pandas as pd

bins = [20, 30, 40, 50, 60]
labels = ["20-30", "30-40", "40-50", "50-60"]

df["AgeGroup"] = pd.cut(df["age"], bins=bins, labels=labels)
```

```python
df.groupby("AgeGroup")["income"].mean()
import matplotlib.pyplot as plt

df.boxplot(column="income", by="AgeGroup")
plt.title("Income Distribution by Age Group")
plt.suptitle("")
plt.xlabel("Age Group")
plt.ylabel("Income")
plt.show()
```

# Pay Equity Assessment

```python
import pandas as pd

# Group by Age and Experience
salary_check = df.groupby(["age", "experience"])["income"].agg(["count", "mean", "min", "max", "std"])

print(salary_check)
```

```python
# Jahan same age & experience ke multiple employees hon
multiple_employees = salary_check[salary_check["count"] > 1]

print(multiple_employees)

multiple_employees["salary_difference"] = multiple_employees["max"] - multiple_employees["min"]

print(multiple_employees)
```

```python

# 1 Average income by age
age_income = df.groupby("age")["income"].mean()

# 2 Average income by experience
exp_income = df.groupby("experience")["income"].mean()

# 3 Line plot: Income vs Age
plt.figure(figsize=(12,5))
plt.plot(age_income.index, age_income.values, marker='o', color='blue')
plt.title("Average Income vs Age")
plt.xlabel("Age")
plt.ylabel("Average Income")
plt.grid(True)
plt.show()

# 4 Line plot: Income vs Experience
plt.figure(figsize=(12,5))
plt.plot(exp_income.index, exp_income.values, marker='o', color='green')
plt.title("Average Income vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Average Income")
plt.grid(True)
plt.show()

```
